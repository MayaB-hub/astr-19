```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100) 
def exp_func(x): 
    return np.exp(x) 
```

```python
y = exp_func(x) 
plt.plot(x,y)
plt.xlabel("Time [miliseconds]")
plt.ylabel("Awesomeness")
plt.title("Exponential Growth of Awesomeness")
plt.grid(True)

plt.savefig("awesomeness_plot.pdf")
plt.show()
```
