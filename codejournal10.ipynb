```python
import numpy as np 
import matplotlib.pyplot as plt
```

pulling 1000 random numbers from numpy from a gamma distribution where k = 2.0 and the scale parameter is 2.0 

```python
data = np.random.gamma(shape=2.0, scale=2.0, size=1000)

plt.figure(figsize=(8,5))
plt.hist(data, bins=100, color='purple', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Gamma Distributed Random Numbers')

plt.savefig('gamma_distribution_historgram.pdf')
plt.show()
```
